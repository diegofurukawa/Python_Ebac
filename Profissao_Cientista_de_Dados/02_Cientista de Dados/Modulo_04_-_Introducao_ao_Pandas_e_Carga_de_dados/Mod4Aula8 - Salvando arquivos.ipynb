# Salvando dataframes em CSV, Excel e JSON

```python
import pandas as pd
import numpy as np
```

```python
np.random.random_sample(size=(100,10))*100
```

```python
df = pd.DataFrame(np.random.random_sample(size=(100,10))*100, 
            columns = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])
```

```python
df.head(1)
```

## Salvando em CSV

```python
df.to_csv('dataframe_aleatorio.csv')
```

```python
df.to_csv('./output/dataframe_aleatorio.csv')
```

```python
df.to_csv('./output/dataframe_aleatorio_index.csv', index= False)
```

```python
df.to_csv('./output/dataframe_aleatorio_index_sep.csv', index= False, sep=';')
```

```python
df.to_csv('./output/dataframe_aleatorio_index_sep_dec.csv', index= False, sep=';', decimal=',')
```

## Salvando em Excel

```python
df.to_excel('./output/dataframe_aleatorio.xlsx')
```

```python
df.to_excel('./output/dataframe_aleatorio_index.xlsx',
            index=False,
            startrow=1,
            sheet_name='aleatorio')
```

## Salvando em JSON

```python
df.to_json('./output/dataframe_aleatorio.json')
```
