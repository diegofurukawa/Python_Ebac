## 1\. Listas

Criei uma lista chamada `filmes` com o nome dos 10 primeiros filmes mais bem avaliados no site no [IMDB](https://www.imdb.com/chart/top/). Imprima o resultado.

```
# Lista para Armazenar os 10 primeiros filmes do IMDB
filmes = [
            'Um Sonho de Liberdade'
            ,'O Poderoso Chefão'
            ,'Batman: O Cavaleiro das Trevas'
            ,'O Poderoso Chefão II'
            ,'12 Homens e uma Sentença'
            ,'A Lista de Schindler'
            ,'O Senhor dos Anéis: O Retorno do Rei'
            ,'Pulp Fiction: Tempo de Violência'
            ,'O Senhor dos Anéis: A Sociedade do Anel'
            ,'Três Homens em Conflito'
          ]

print(filmes)
print(type(filmes))
```

Simule a movimentação do *ranking*. Utilize os métodos `insert` e `pop` para trocar a posição do primeiro e do segundo filme da lista. Imprima o resultado.



```
# Print apenas para validar a sequencia inicial da tarefa anterior
print(filmes)

# Removendo o primeiro filme
filmes.pop(0)
print(filmes)

# Inserindo o filme removido na nova posicao
filmes.insert(1, 'Um Sonho de Liberdade')
print(filmes)

```

## 2\. Conjuntos

Aconteceu um erro no seu *ranking*. Simule a duplicação dos três últimos filmes da lista. Imprima o resultado.

```
# Lista para Armazenar os 10 primeiros filmes do IMDB
filmes_ex2 = [
            'Um Sonho de Liberdade'
            ,'O Poderoso Chefão'
            ,'Batman: O Cavaleiro das Trevas'
            ,'O Poderoso Chefão II'
            ,'12 Homens e uma Sentença'
            ,'A Lista de Schindler'
            ,'O Senhor dos Anéis: O Retorno do Rei'
            ,'Pulp Fiction: Tempo de Violência'
            ,'O Senhor dos Anéis: A Sociedade do Anel'
            ,'Três Homens em Conflito'
          ]

print(filmes_ex2)
# print(type(filmes))

# Inserindo o filme para duplicar na 8 posicao 
filmes_ex2.insert(7, 'Pulp Fiction: Tempo de Violência')
# Inserindo o filme para duplicar na 9 posicao 
filmes_ex2.insert(9, 'O Senhor dos Anéis: A Sociedade do Anel')
# Inserindo o filme para duplicar na 10 posicao 
filmes_ex2.insert(11, 'Três Homens em Conflito')
print(filmes_ex2)
```

Utiliza a conversão `set` e `list` para remover os valores duplicados. Imprima o resultado. 

```
filmes_list_ex2 = set(filmes_ex2)
print(list(filmes_list_ex2))
```
