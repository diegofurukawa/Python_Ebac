## 2\. Funções para arquivo txt

Complete a função abaixo para extrair uma as palavras de uma linha do arquivo txt em uma lista.

```python
text_lines = []
words_line = []
lines = []
number_line = 0


with open(file ='./musica.txt', mode='r', encoding='utf8') as arquivo:
	#linha = arquivo.readline() # lê o cabeçalho
	linha = arquivo.readline() # lê a primeira linha
	while linha:
		number_line = number_line + 1
		line_separated = linha.split(sep='\n')
		text_lines.append(line_separated) # salva o valor na lista
		if number_line == 10:
			word_separated = linha.split(sep=' ') # quebra a string por espacos e salva os resultados em uma lista
			words_line.append(word_separated)
		lines.append(number_line) # salva o valor na lista
		linha = arquivo.readline() # lê uma nova linha, se a linha não existir, salva o valor None

	print(text_lines)
	print(lines)
	print(words_line)
	print("Número de linhas é: ", number_line)
```

```python
def extrai_linha_txt(nome_arquivo: str, numero_linha: int) -> str:

  text_lines = []
  words_line = []
  lines = []
  number_line = 0

  # leia o arquivo com o comando 'with' utilizando o parametro 'nome_arquivo'
  # extraia a linha do arquivo utilizando o parametro 'numero_linha'
  # quebre a linha em palavras com o comando split, note que o separador é um espaço ' '
  
  with open(file =nome_arquivo, mode='r', encoding='utf8') as arquivo:
    #linha = arquivo.readline() # lê o cabeçalho
    linha = arquivo.readline() # lê a primeira linha
    while linha:
      number_line = number_line + 1
      line_separated = linha.split(sep='\n')
      text_lines.append(line_separated) # salva o valor na lista
      if number_line == numero_linha:
        word_separated = linha.split(sep=' ') # quebra a string por espacos e salva os resultados em uma lista
        words_line.append(word_separated)
      lines.append(number_line) # salva o valor na lista
      linha = arquivo.readline() # lê uma nova linha, se a linha não existir, salva o valor None

  return words_line
```
