## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
ucancao=(cancao.upper())
print(ucancao)

fcancao = (cancao.find(','))
print(fcancao)

rcancao = (cancao.replace(', ', ' - '))
print(rcancao)

```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'


#Fatiamento Fixo
fvarselic=noticia.find('%')
fvarselic=(fvarselic-4)
selic=noticia[fvarselic:fvarselic+5]
#print(fvar)
#print('A taxa da Selic e: ' + noticia[fvarselic]+noticia[fvarselic+1]+noticia[fvarselic+2]+noticia[fvarselic+3]+noticia[fvarselic+4])
print('A taxa da Selic e: ' + selic)

fvarano=noticia.find('6')
ano=noticia[fvarano]
#print(fvar)
#print('A taxa da Selic e: ' + noticia[fvarselic]+noticia[fvarselic+1]+noticia[fvarselic+2]+noticia[fvarselic+3]+noticia[fvarselic+4])
print('O ano e: ' + ano)


```
