####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# Montando DataFrame para Grafico

sinasc_barras = sinasc[['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']]
display(sinasc_barras)
```

```python
# sinasc_barras.plot.bar()

# 'APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE'
sinasc_barras['APGAR1'].value_counts().sort_index().plot.bar()
```

```python

# 'APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE'
sinasc_barras['APGAR5'].value_counts().sort_index().plot.bar()
```

```python

# 'APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE'
sinasc_barras['SEXO'].value_counts().sort_index().plot.bar()

```

```python

# 'APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE'
sinasc_barras['GESTACAO'].value_counts().sort_index().plot.bar()
```

```python

# 'APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE'
sinasc_barras['ESTCIVMAE'].value_counts().sort_index().plot.bar()
```

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(3, 2)

# [['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
sns.countplot(ax = axes[0, 0], x='APGAR1', data=sinasc_barras)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=sinasc_barras)
sns.countplot(ax = axes[1, 0], x='SEXO', data=sinasc_barras)
sns.countplot(ax = axes[1, 1], x='GESTACAO', data=sinasc_barras)
sns.countplot(ax = axes[2, 0], x='ESTCIVMAE', data=sinasc_barras)

```

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

# [['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']]
sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc_barras)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinasc_barras)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc_barras)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc_barras)

```
