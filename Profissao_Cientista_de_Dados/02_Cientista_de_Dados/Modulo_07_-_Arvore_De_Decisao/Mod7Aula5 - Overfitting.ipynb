```python
x1 = np.array([10,22,13,42,25,36,24])
x2 = np.array([100,200,300,400,500,600,400])
y = np.array([0,0,0,1,0,1,1])
```

```python
teste = pd.DataFrame({'x1': x1, 'x2':x2})
```

```python
clf = DecisionTreeClassifier(random_state=100)
clf
```

```python
clf = clf.fit(teste,y)
clf
```

```python
plt.figure(figsize=(25, 10))
plot_tree(clf,
          filled=True,
          class_names=['não', 'sim'],
          feature_names=['x1','x2']);
```
